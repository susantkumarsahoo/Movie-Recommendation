```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['genres'].values
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L        
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

```python
movies['cast'].values
```

```python
def convert3(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
#movies similarity distance
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
#fetching movies name to index number
new_df[new_df['title'] == 'Spectre'].index[0]
```

```python
#movies similarity recommend
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#movies fetching index to movies name
new_df.iloc[1537].title
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('John Carter')
```
